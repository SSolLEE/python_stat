## 1.2 변수의 종류
## 1.2.1 질적 변수와 양적 변수
- 질적 변수: 선택이 필요한 변수(좋음, 보통, 나쁨), 종류를 구별하기 위한 변수(혈액형)
- 양적 변수: 양을 표현하는 변수(시험 점수, 신장)
- 수치 변수라고 해서 모두 양적 변수인 것은 아니다(성별을 데이터상에서 쉽게 처리하기 위해 남성을 0, 여성을 1과 같이 표시하는 경우가 있음.)

## 1.2.2 척도 수준
- 질적 변수: 명의/순서 척도, 양적 변수: 간격/비례 척도
- 명의 척도: 단순 분류, 변수의 동일성 여부 / 학생번호, 성별
- 순서 척도: 순서 관계, 대소 관계 / 성적 순위, 설문조사의 만족도
- 간격 척도: 대소 관계 + 차이 / 연도, 온도
- 비례 척도: 대소 관계, 차이, 비 / 길이, 무게

## 1.2.3 이산형 변수와 연속형 변수
- 이산형 변수: 0, 1, 2, ...와 같이 하나하나의 값을 취하는 변수. 인접한 숫자 사이에 값이 존재하지 않음 / 학교를 결석한 횟수
- 연속형 변수: 연속적인 값을 취할 수 있는 변수, 어떤 두 숫자 사이에도 반드시 숫자가 존재 / 길이, 무게, 시간
